CONCEPTOS BÁSICOS

### GRAFO
------------------------------
*   Un **grafo** es un diagrama que consiste de puntos llamados **vertices**, 
conectados por líneas, llamadas **aristas** (edges); cada arista conecta exactamente dos vertices.



**Figura 1**. Grafo [1]

Un grafo puede ser utilizado para representar cualquier tipo de información que pueda ser modelada como objetos y que existe una relación entre estos objetos.

----------------------------

### DIGRAFO
*   Un **digrafo** es un diagrama que consiste de puntos, llamados vertices, conectados por lineas directoras, llamadas **arcos**, cada arco conecta exactamente dos vertices.


**Figura 2**. Digrafo [1].

----------------------------------

Los problemas en teoría de grafos incluyen colorear grafos, encontrar un camino entre dos estados o nodos, o encontrar el camino más corto a través de un grafo, entre muchos otros.

### Notación
Representar un grafo se puede hacer de varias maneras, la forma correcta de representarlos del algoritmo que se implemente. 

Un grafo puede representarse cómo:

\begin{equation}
G = (V,E)
\end{equation}

Este está definido por un conjunto de vértices, denominados V, y un conjunto de aristas, denominado E. El conjunto de aristas son subconjuntos de V donde cada miembro de E tiene cardinalidad 2. 

Considere el grafo no dirigido de la siguiente imagen.



Los conjuntos $V = \{A, E, G, L, W\}$ y $E = \{\{A, E\},\{A, W\},\{G, L\},\{G, W\},\{L, W\}\}$ definen este grafo. Dado que cada arista es en sí misma un conjunto de cardinalidad 2, en un grafo no dirigido, el orden de los vértices en cada conjunto de aristas no importa. Por ejemplo, $\{A, El\}$ es la misma arista que $\{El, A\}$.
