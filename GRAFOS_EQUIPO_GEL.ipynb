## HISTORIA

La primera mencion de los grafos fue hecha por el matemático suizo Leonhard Euler, en una carta dirigida a la Academia de San Petersburgo, en 1736, en ella se aborda la solucion del problema de los puentes de Königsberg.

El problema es el siguiente: 

¿Es posible dar un paseo comenzando desde cualquiera de estas regiones, pasando por todos los puentes, recorriendo una sola vez cada uno y regresando al mismo lugar?

## INTRODUCCIÓN A GRAFOS
CONCEPTOS BÁSICOS

### Grafo 
*   Un **grafo** es un diagrama que consiste de puntos llamados **vertices**, 
conectados por líneas, llamadas **aristas** (edges); cada arista conecta exactamente dos vertices.

<div align="center">

</div>

<center>  <b>Figura 2. </b>  Grafo [1]. </center>



Un grafo puede ser utilizado para representar cualquier tipo de información que pueda ser modelada como objetos y que existe una relación entre estos objetos.



### Digrafo
*   Un **digrafo** es un diagrama que consiste de puntos, llamados vertices, conectados por lineas directoras, llamadas **arcos**, cada arco conecta exactamente dos vertices.

<div align="center">

</div>

<center>  <b>Figura 3. </b>  Digrafo [1]. </center>

Tanto los grafos dirigidos como los no dirigidos se pueden usar para modelar muchos problemas.

## GRAFOS EN MATEMÁTICAS

La teoría de grafos se usa en matemáticas discretas y en ciencias de la computación, se relaciona mucho con teoría de conjuntos. 

> Un conjunto es una colección desordenada de elementos.

> Un subconjunto es una colección de elementos de su superconjunto.

Un **gráfo** está definido por un conjunto de vértices, denotados por $V$, y un conjunto de aristas denotados por $E$. El conjunto de aristas son subconjuntos de $V$ donde cada miembro de $E$ tiene cardinalidad 2.

\begin{equation}
G = (V,E)
\end{equation}


**EJEMPLO 1:** Considere el grafo no dirigido de la siguiente imagen.

<div align="center">

<br>

Los conjuntos $V = \{A, E, G, L, W\}$ y $E = \{\{A, E\},\{A, W\},\{G, L\},\{G, W\},\{L, W\}\}$ definen este grafo. Dado que cada arista es en sí misma un conjunto de cardinalidad 2, en un grafo no dirigido, el orden de los vértices en cada conjunto de aristas no importa. Por ejemplo, $\{A, E\}$ es la misma arista que $\{E, A\}$.

Un grafo dirigido G = (V,E) se define de la misma forma que uno no dirigido excepto que el conjunto de aristas, **E**, es un conjunto de tuplas en lugar de subconjuntos.

### Grafos ponderados
Un grafo con peso es aquel en el que las aristas (edges) tiene un valor definido.

$G = (V,E,W)$

Donde $V=\{v_{1},...,v_{n}\}$ es un conjunto de vertices, $E=\{e_{1},...,e_{n}\}$ es un conjunto de aristas y $W=\{w_{1},...,w_{n}\}$ es un conjunto de pesos que estan asociadas a cada arista.

## ALGORITMOS DE GRAFOS

### ALGORITMO BFS
El algoritmo de búsqueda en anchura (**Breadth-First Search**) realiza una exploración donde dado un nodo, recorremos todos los vecinos primero. 

<div align="center">

</div>

<center>  <b>Figura . </b>  Algoritmo BFS. </center>

La complejidad temporal de este algoritmo es de $O(V+E)$, eso significa que se ejecuta en tiempo lineal con respecto a la candidad de nodos y aristas del grafo.
Para llegar al nodo actual encuentra el caminino mínimo de cada nodo al actual en grafos no ponderados. Es decir, en grafos donde desplazarse no implica ningún costo.


### ALGORITMO DFS
Una búsqueda en profundidad (**Depth First Search**) es un algoritmo de búsqueda que recorre los nodos de un grafo de manera que primero primero va expandiendo cada uno de los nodos que va localizando, de forma recurrente (desde el nodo padre hacia el nodo hijo). Es decir vamos al fondo de la primera rama, antes de pasar a la siguiente.

<div align="center">

</div>

<center>  <b>Figura . </b>  Algoritmo DFS. </center>

La complejidad temporal de este algoritmo también es de $O(V+E)$

### ALGORITMO DIJKTRA

El algoritmo de Dijkstra es famoso en Ciencias de la Computación y tiene muchas aplicaciones, desde redes hasta planificación de la construcción.
El algoritmo de Dijkstra, también llamado algoritmo de caminos mínimos, es un algoritmo para la determinación del camino más corto entre un vértice origen, a cualquier otro vertice en el grafo ponderado. 

La complejidad de este algoritmo es $O(n^2)$, donde n es el número de vértices. Este algoritmo puede implementarse en problemas de planificación; tomando a los nodos del grafo como metas temporales y los vértices como las transiciones para llegar a cada una de estas metas hasta llegar al nodo deseado.
