# Graphs
Graphs can be used to represent information that can be shown as __vertices__ and __edges__, and most importantly, it exist a relationship between them

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
```

<code>G = nx.Graph()</code> By definition, Graph is a collection of nodes(aka Vertices) along with identified pair of nodes (aka edges, links). In particular, Graph is used for __UNDIRECTED__ Graph. If we want to use a _directed_ graph, we must use <code>G = nx.DiGraph()</code>

## Creating an instance of a Graph

```python
G = nx.Graph()
```

## Adding nodes to an existing instance

We can add __nodes__ to an existing graph. E.g., we can add one node at the time to our G graph instance: <code>G.add_node(1)</code>

```python
G.clear()
G.add_node(1)
```

We can also add several nodes from a list data type: <code>G.add_nodes_from([x, y, z])</code>

```python
G.add_nodes_from([2, 3, 4])
nx.draw(G, with_labels=True)
```

```python
G.add_edge(1,2)
nx.draw(G, with_labels=True)
```

If we try to add multitple edges to connect our nodes, by using the <code>G.add_node()</code> method. We wil get an error

```python
G.add_edge[(1,2), (1,3)]
```

For adding several edges, we must use the <code>G.add_edges_from([(x, y), (z,a)...,(q,w)])

```python
G.add_edges_from([(1,3), (1,2)])
nx.draw(G, with_labels=True)
```

### Views

```python
print(f"The number of nodes are: {G.number_of_nodes()}")
print(f"The number of edges are: {G.number_of_nodes()}")
print(f"G.degree = {G.degree}")
```

## Example 1

```python
example1 = nx.Graph()
example1.add_nodes_from([0, 1, 2, 3])
example1.add_edges_from([(1,2), (2,3), (3,1), (0,1)])
print(f"The number of nodes are: {example1.number_of_nodes()}")
print(f"The number of edges are: {example1.number_of_nodes()}")
print(f"example1.degree = {example1.degree}")
print(f"example1.adj = {example1.adj}")
nx.draw(example1, with_labels=True)
```

```python
class Vertice:
    # constructor
    def __init__(self, i):
        self.id = i
        self.visitado = False
        self.nivel = -1
        self.vecinos = []

    def agregarVecino(self, v):
        if v not in self.vecinos:
            self.vecinos.append(v)

class Grafica:
    def __init__(self):
        self.vertices = {}

    def agregaVertice(self,v):
        if v not in self.vertices:
            self.vertices[v] = Vertice(v)

    def agregarArista(self,a,b):
        if a in self.vertices and b in self.vertices:
            self.vertices[a].agregarVecino(b)
            self.vertices[b].agregarVecino(a)

def main():

    g = Grafica()

    l = [0,1,2,3,4,5,6]
    for v in l:
        g.agregaVertice(v)

    l = [2,0,0,6,6,3,0,5,6,5,0,1,6,4,1,4]
    for i in range(0,len(l)-1,2):
        g.agregarArista(l[i],l[i+1])

    for v in g.vertices:
        print(v,g.vertices[v].vecinos)

main()
```
