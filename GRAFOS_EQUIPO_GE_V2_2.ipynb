## HISTORIA

La primera mencion de los grafos fue hecha por el matemático suizo Leonhard Euler, en una carta dirigida a la Academia de San Petersburgo, en 1736, en ella se aborda la solucion del problema de los puentes de Königsberg.

El problema es el siguiente: 

¿Es posible dar un paseo comenzando desde cualquiera de estas regiones, pasando por todos los puentes, recorriendo una sola vez cada uno y regresando al mismo lugar?

## INTRODUCCIÓN A GRAFOS
CONCEPTOS BÁSICOS

### Grafo 
*   Un **grafo** es un diagrama que consiste de puntos llamados **vertices**, 
conectados por líneas, llamadas **aristas** (edges); cada arista conecta exactamente dos vertices.

<div align="center">

</div>

<center>  <b>Figura 2. </b>  Grafo [1]. </center>



Un grafo puede ser utilizado para representar cualquier tipo de información que pueda ser modelada como objetos y que existe una relación entre estos objetos.



### Digrafo
*   Un **digrafo** es un diagrama que consiste de puntos, llamados vertices, conectados por lineas directoras, llamadas **arcos**, cada arco conecta exactamente dos vertices.

## GRAFOS EN MATEMÁTICAS

La teoría de grafos se usa en matemáticas discretas y en ciencias de la computación, se relaciona mucho con teoría de conjuntos. 

> Un conjunto es una colección desordenada de elementos.

> Un subconjunto es una colección de elementos de su superconjunto.

Un **gráfo** está definido por un conjunto de vértices, denotados por $V$, y un conjunto de aristas denotados por $E$. El conjunto de aristas son subconjuntos de $V$ donde cada miembro de $E$ tiene cardinalidad 2.

\begin{equation}
G = (V,E)
\end{equation}


**EJEMPLO 1:** Considere el grafo no dirigido de la siguiente imagen.

<div align="center">

<br>

Los conjuntos $V = \{A, E, G, L, W\}$ y $E = \{\{A, E\},\{A, W\},\{G, L\},\{G, W\},\{L, W\}\}$ definen este grafo. Dado que cada arista es en sí misma un conjunto de cardinalidad 2, en un grafo no dirigido, el orden de los vértices en cada conjunto de aristas no importa. Por ejemplo, $\{A, E\}$ es la misma arista que $\{E, A\}$.

Un grafo dirigido G = (V,E) se define de la misma forma que uno no dirigido excepto que el conjunto de aristas, **E**, es un conjunto de tuplas en lugar de subconjuntos.

### Grafos ponderados

## ALGORITMOS DE GRAFOS

### ALGORITMO BFS
El algoritmo de búsqueda en anchura (**Breadth-First Search**) realiza una exploración donde dado un nodo, recorremos todos los vecinos primero. 

<div align="center">

</div>

<center>  <b>Figura 12. </b>  Algoritmo BFS. </center>

La complejidad temporal de este algoritmo es de $O(V+E)$, eso significa que se ejecuta en tiempo lineal con respecto a la candidad de nodos y aristas del grafo.
Para llegar al nodo actual encuentra el caminino mínimo de cada nodo al actual en grafos no ponderados. Es decir, en grafos donde desplazarse no implica ningún costo.


### ALGORITMO DFS
Una búsqueda en profundidad (**Depth First Search**) es un algoritmo de búsqueda que recorre los nodos de un grafo de manera que primero primero va expandiendo cada uno de los nodos que va localizando, de forma recurrente (desde el nodo padre hacia el nodo hijo). Es decir vamos al fondo de la primera rama, antes de pasar a la siguiente.

<div align="center">

</div>

<center>  <b>Figura 13. </b>  Algoritmo DFS. </center>

La complejidad temporal de este algoritmo también es de $O(V+E)$

### ALGORITMO DIJKTRA

El algoritmo de Dijkstra es famoso en Ciencias de la Computación y tiene muchas aplicaciones, desde redes hasta planificación de la construcción.
El algoritmo de Dijkstra, también llamado algoritmo de caminos mínimos, es un algoritmo para la determinación del camino más corto entre un vértice origen, a cualquier otro vertice en el grafo ponderado. 

La complejidad de este algoritmo es $O(n^2)$, donde n es el número de vértices. Este algoritmo puede implementarse en problemas de planificación; tomando a los nodos del grafo como metas temporales y los vértices como las transiciones para llegar a cada una de estas metas hasta llegar al nodo deseado [3].

<div align="center">

</div>

<center>  <b>Figura 14. </b>  Algoritmo DIKKSTRA. </center>

### ALGORITMO ESTRELLA
El algoritmo A* es un algoritmo de busqueda que encuentra el camino más corto en un problema que puede ser modelado como un Grafo. Dados un nodo de inicio y con un nodo final o de destino. El algoritmo tipo A* NO es un algoritmo __Blind__.

Un algoritmo __Blind__ NO toma en cuenta los __PESOS__ de las aristas.

La palabra clave en el algoritmo A* es la función __heuristic__. La función __heuristic__ 'estima' el costo del nodoo __actual__ al nodo __final__ (destino) usando una métrica como la distancia __Manhattan__ (Distancia Euclidiana) [6].La complejidad del algoritmo dependerá de la función heuristca, $O(b^d)$

Función __heuristica__ $f(n) = g(n) + h(n)$


* $f(n)$ -> Calculated total cost of path
* $g(n)$ -> The cost of path between the first node and the current node
* $h(n)$ -> Heuristic function

<div align="center">

</div>

<center>  <b>Figura 15. </b>  Algoritmo A*. </center>

## Costo
Camino $X -> B:$ $f(n) = g(n) + h(n) = 1 + 4 = 5$ \
Camino $X -> A:$ $f(n) = g(n) + h(n) = 5 + 1 = 6$ (hold) \
Camino $X -> B -> C:$ $f(n) = g(n) + h(n) = 1 + 2 + 2 = 5$ \
Camino $X -> A -> Y:$ $f(n) = g(n) + h(n) = 5 + 1 + 0 = 6$ (hold) \
Camino $X -> B -> C -> Y:$ $f(n) = g(n) + h(n) = 1 + 2 + 2 + 0= 5$ \
Camino con menor costo:  $X -> B -> C -> Y$

## Weighted Graps

```python
fig, ax = plt.subplots(figsize = (16, 10), dpi = 200)
example3 = nx.Graph()
example3.add_nodes_from(['Luis','Antonio','Gabriela','Eli','William'])
example3.add_edge('Luis','Gabriela', weight=0.4)
example3.add_edge('Eli','Antonio',weight=0.5)
example3.add_edge('William','Antonio',weight=0.6)
example3.add_edge('William', 'Gabriela',weight = 0.9)
example3.add_edge('William','Luis A',weight=0.9)

print(f"The number of nodes are: {example3.number_of_nodes()}")
print(f"The number of edges are: {example3.number_of_nodes()}")
print(f"example1.degree = {example3.degree}")
print(f"example1.adj = {example3.adj}")

elarge = [(u, v) for (u, v, d) in example3.edges(data=True) if d["weight"] > 0.5]
esmall = [(u, v) for (u, v, d) in example3.edges(data=True) if d["weight"] <= 0.5]

pos = nx.spring_layout(example3, seed=7) 
nx.draw_networkx_nodes(example3, pos, node_size=600)
nx.draw_networkx_edges(example3, pos, edgelist=elarge, width=6)
nx.draw_networkx_edges(
    example3, pos, edgelist=esmall, width=6, alpha=0.5, edge_color="b", style="dashed"
)
nx.draw_networkx_labels(example3, pos, font_size=15, font_family="sans-serif")
# edge weight labels
edge_labels = nx.get_edge_attributes(example3, "weight")
nx.draw_networkx_edge_labels(example3, pos, edge_labels)
#nx.draw(example2, with_labels=True)
```

```python
class Vertice:
    # constructor
    def __init__(self, i):
        self.id = i
        self.visitado = False
        self.nivel = -1
        self.vecinos = []

    def agregarVecino(self, v):
        if v not in self.vecinos:
            self.vecinos.append(v)

class Grafica:
    def __init__(self):
        self.vertices = {}

    def agregaVertice(self,v):
        if v not in self.vertices:
            self.vertices[v] = Vertice(v)

    def agregarArista(self,a,b):
        if a in self.vertices and b in self.vertices:
            self.vertices[a].agregarVecino(b)
            self.vertices[b].agregarVecino(a)

def main():

    g = Grafica()

    l = [0,1,2,3,4,5,6]
    for v in l:
        g.agregaVertice(v)

    l = [2,0,0,6,6,3,0,5,6,5,0,1,6,4,1,4]
    for i in range(0,len(l)-1,2):
        g.agregarArista(l[i],l[i+1])

    for v in g.vertices:
        print(v,g.vertices[v].vecinos)

main()
```
